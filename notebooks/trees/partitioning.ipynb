## Classification

```python
from dtreeviz.trees import *

iris = load_iris()
X = iris.data
y = iris.target
len(X), len(y)

feature_names = iris.feature_names
class_names = list(iris.target_names)
print(feature_names)
figsize = (6,2)
X = X[:,2]
fig, ax = plt.subplots(1, 1, figsize=figsize)
ct = ctreeviz_univar(ax, X, y, max_depth=1,
                     feature_name = 'petal length (cm)', class_names=class_names,
                     target_name='iris',
                     nbins=40, gtype='barstacked',
                     show={'splits','title'})
plt.tight_layout()
plt.savefig("/Users/parrt/github/msds621/projects/dtree/images/iris-1-bar.svg")
plt.show()
```
